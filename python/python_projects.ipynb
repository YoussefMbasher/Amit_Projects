```python
class bank_account:
    def __init__(self, name, balance):
        self.name=name
        self.balance=balance
    
    def deposit(self, amount):
        self.balance+=amount
        print(f'Deposited {amount}. New balance: {self.balance}')
        
    def withdraw(self, amount):
        if amount > self.balance:
            print('Insufficient funds')
        else:
            self.balance -= amount
            print(f'Withdrew {amount}. New balance: {self.balance}')
            
    def check_balance(self):
         print(f'Current balance: {self.balance}')

        
        
acc1=bank_account('John', 1000)
acc1.check_balance()
acc1.deposit(500)
acc1.withdraw(200)
acc1.check_balance()
```

```python
class Calculator:
    def __init__(self):
        """Initializes the calculator class."""
        

    def add(self, a, b):
        """Add two numbers."""
        print(a+b)

    def subtract(self, a, b):
        """Subtract two numbers."""
        print(a-b)

    def multiply(self, a, b):
        """Multiply two numbers."""
        print(a*b)

    def divide(self, a, b):
        """Divide two numbers."""
        print(a/b)

    def calculator(self):
        """Main calculator logic to get user input and perform operations."""
        print("Welcome to the calculator!")
        print("Choose an operation:")
        print("1: Add")
        print("2: Subtract")
        print("3: Multiply")
        print("4: Divide")

        choice = input("Enter the number of the operation you want to perform (1/2/3/4): ")

        if choice in ['1', '2', '3', '4']:
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))
            except ValueError:
                print("Invalid input! Please enter numbers only.")
                return
            
            if choice == '1':
                print(f"{num1} + {num2} = {self.add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {self.subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {self.multiply(num1, num2)}")
            elif choice == '4':
                result = self.divide(num1, num2)
                if isinstance(result, str):
                    print(result)  # Prints error message
                else:
                    print(f"{num1} / {num2} = {result}")
        else:
            print("Invalid choice! Please select a valid operation (1/2/3/4).")

# Usage example
if __name__ == "__main__":
    calculator = Calculator()
    calculator.calculator()
```

```python
#task 4
class TextFileReader:
    def __init__(self, file_path):
        self.file_path = file_path
        self.content = None
    
    def read_file(self):
        """Read the file and store its contents."""
        try:
            with open(self.file_path, 'r', encoding='utf-8') as file:
                self.content = file.read()
        except FileNotFoundError:
            print(f"Error: The file at {self.file_path} was not found.")
        except Exception as e:
            print(f"An error occurred: {e}")

    def count_lines(self):
        """Return the number of lines in the file."""
        if self.content is not None:
            return len(self.content.splitlines())
        return 0

    def count_words(self):
        """Return the total number of words in the file."""
        if self.content is not None:
            return len(self.content.split())
        return 0

    def count_characters(self):
        """Return the total number of characters in the file."""
        if self.content is not None:
            return len(self.content)
        return 0

    def display_content(self):
        """Print the content of the file."""
        if self.content is not None:
            print(self.content)
        else:
            print("No content to display. Please read the file first.")

# Example usage:
if __name__ == "__main__":
    file_path = "Department.txt"  
    text_reader = TextFileReader(file_path)

    text_reader.read_file()
    print(f"Lines: {text_reader.count_lines()}")
    print(f"Words: {text_reader.count_words()}")
    print(f"Characters: {text_reader.count_characters()}")
    text_reader.display_content()

    
```

```python
#task 5
def count_word_frequencies(words):
    """Count the frequency of each unique word in a list of strings."""
    frequency = {}
    
    for word in words:
        # Convert to lowercase for case-insensitive counting if desired
        word = word.strip()  
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
            
    return frequency

# Example usage:
if __name__ == "__main__":
    words_list = ["Welcome", "youssef", "Hi", "youssef", "Welcome", "Hi", "No", "youssef", "No", "youssef"]
    result = count_word_frequencies(words_list)
    print(result)  
```

```python
#task 6
def read_txt_file(file_path):
    """Read the contents of a specified text file and return it as a string."""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()
    except FileNotFoundError:
        return "Error: The specified file was not found."
    except IOError:
        return "Error: An IO error occurred while reading the file."
    except Exception as e:
        return f"An unexpected error occurred: {e}"

class UserExtractor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.usernames = {}

    def extract_usernames(self):
        """Extract usernames from the file and store them in a dictionary."""
        file_content = read_txt_file(self.file_path)
        
        if isinstance(file_content, str) and file_content.startswith("Error:"):
            return file_content 

        for line in file_content.splitlines():
            if ':' in line: 
                username, _ = line.split(':', 1)  
                username = username.strip()  
                if username:  
                    self.usernames[username] = True  

        return self.usernames

# Example usage:
if __name__ == "__main__":
    file_path = 'Department.txt'  # Replace with your valid text file path
    user_extractor = UserExtractor(file_path)
    
    result = user_extractor.extract_usernames()
    print(result)  # Output: Dictionary of unique usernames or an error message
```
